## ADD NAMED EDGE

```python
from cmath import nan
def add_nammed_edges(nodes_data):

        stmt_type=['Identifier','FunctionDeclaration', 'ArrowFunctionExpression', 'FunctionExpression', 'CallExpression', 'NewExpression']
        
        nodes_data = nodes_data[nodes_data.type.isin(stmt_type)]
        call_sites = nodes_data[((nodes_data['type']=='CallExpression')|(nodes_data['type']=='NewExpression'))]

        function_defs = nodes_data[nodes_data.type.isin(['FunctionDeclaration', 'ArrowFunctionExpression', 'FunctionExpression'])]

        identifiers = nodes_data[(nodes_data['type']=='Identifier')]
        id_dict={}

        for i in range(len(identifiers)):
                id = identifiers.iloc[i]['new_id']
                key = identifiers.iloc[i]['name']
                id_dict[key]=id

        u=[]
        v=[]
        for i in range(len(call_sites)):
            id = call_sites.iloc[i]['new_id']
            name = call_sites.iloc[i]['name']
            if str(name)!='nan' and name in id_dict:
                nammed_id = id_dict[name]
                # print(nammed_id)
                u.append(id)
                v.append(nammed_id)
        
        new_df = pd.DataFrame({'src':u,'dst':v})
        # print(new_df)

        u=[]
        v=[]
        for i in range(len(function_defs)):
            id = function_defs.iloc[i]['new_id']
            name = function_defs.iloc[i]['name']
            if str(name)!='nan' and name in id_dict:
                nammed_id = id_dict[name]
                # print(nammed_id)
                u.append(id)
                v.append(nammed_id)
        
        func_df = pd.DataFrame({'src':u,'dst':v})
        # print(func_df)

        combined_df = pd.concat([new_df,func_df]).reset_index().drop(["index"], axis=1)
        # print(combined_df)

        return combined_df

```
